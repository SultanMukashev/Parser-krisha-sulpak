```python
import pandas as pd
from faker import Faker
import random
from datetime import datetime, timedelta

fake = Faker()

# Set seed for reproducibility
random.seed(42)

# Create a list of Employee data
employees_data = []

for i in range(1, 501):
    employee_id = i
    first_name = fake.first_name()
    last_name = fake.last_name()
    phone_number = f"+7777{random.randint(1000000, 9999999)}"
    start_date = fake.date_between(start_date='-5y', end_date='today')
    end_date = fake.date_between(start_date=start_date, end_date='today')
    shop_id = random.randint(1, 25)
    salary = random.randint(70000, 380000)

    employee_data = [employee_id, first_name, last_name, phone_number, start_date, end_date, shop_id, salary]
    employees_data.append(employee_data)

# Create a DataFrame
columns = ['EmployeeID', 'First_Name', 'Last_Name', 'Phone_Number', 'Start_Date', 'End_Date', 'ShopID', 'Salary']
employees_df = pd.DataFrame(employees_data, columns=columns)

# Save to CSV
employees_df.to_csv('employees_data.csv', index=False)
```

```python
from LxmlSoup import LxmlSoup as lx
import requests
import pandas as pd
url = 'https://www.sulpak.kz/f/smartfoniy?page='
i = 1
res = {'Category':[],"Brand":[],'Model':[],'Price':[]}
def parse(url):
    html = requests.get(url).text
    soup = lx(html)
    phones = soup.find_all('div',class_='product__item product__item-js tile-container')
    for phone in phones:
      category = phone.get('data-list')
      name = ' '.join(phone.get("data-name").split()[2:])
      price = phone.get('data-price')
      brand = phone.get('data-brand')
      res['Category'].append(category)
      res['Brand'].append(brand)
      res['Model'].append(name)
      res["Price"].append(price)
for i in range(30):
    parse(url+str(i))

df = pd.DataFrame(data = res)
print(df)
df.to_excel('phones.xlsx')
```
